```python
import os ; import sys
sys.path.append(os.getcwd()+"/../../src")

from qtcipy import interpolate
```

```python


import numpy as np

def f(x,y):
#    return np.cos(30*x)/(1.+x**2)*(1+(x*y)**2)*np.sin(y)
    r = np.sqrt(x**2 + y**2) 
    return 1./(0.01+(2.-r)**2) + 1./(0.01+(1.-r + x)**2) 


xlim = [-4,4]
ylim = [-4,4]

# create the interpolator
IP = interpolate.Interpolator(f,tol=1e-1,nb=10,xlim=xlim,ylim=ylim,dim=2)



# check the evaluated points as a reference
rse,zse = IP.get_evaluated()
print(len(rse),"evaluations")
rse = np.array(rse)


# evaluate the function and the interpolation
nn = 100
xs = np.linspace(-3,3,nn)
ys = np.linspace(-3,3,nn)
zs,zsi = [],[]
for x in xs:
    for y in ys:
        zs.append(f(x,y))
        zsi.append(IP(x,y))


import matplotlib.pyplot as plt
plt.subplot(1,3,1)
plt.imshow(np.array(zs).reshape(nn,nn).T)

plt.subplot(1,3,2)
plt.imshow(np.array(zsi).reshape(nn,nn).T)

plt.subplot(1,3,3)
plt.scatter(rse[:,0],rse[:,1],c=zse,s=0.1)
plt.axis("equal")

plt.tight_layout()


plt.show()


```
