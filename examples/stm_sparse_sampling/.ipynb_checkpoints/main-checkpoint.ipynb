# Sparse sampling with STM
Let us the QTCI algorithm to perform sparse sampling

```python
import cv2

# Read the image
image = cv2.imread('stm_moire.png')
#image = cv2.imread('stm.png')

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print(image.shape)

import matplotlib.pyplot as plt

plt.imshow(image)
```

```python
import numpy as np

def f(x, y):
    # Get the dimensions of the image
    height, width = image.shape

    # Scale x and y to the image dimensions
    x_scaled = x * (width - 1)
    y_scaled = y * (height - 1)

    # Get the coordinates of the surrounding pixels
    x0 = int(np.floor(x_scaled))
    x1 = min(x0 + 1, width - 1)
    y0 = int(np.floor(y_scaled))
    y1 = min(y0 + 1, height - 1)

    # Calculate the interpolation weights
    a = x_scaled - x0
    b = y_scaled - y0

    # Interpolate the pixel value
    interpolated_value = (image[y0, x0] * (1 - a) * (1 - b) +
                          image[y0, x1] * a * (1 - b) +
                          image[y1, x0] * (1 - a) * b +
                          image[y1, x1] * a * b)

    return interpolated_value


def f(x,y):
    k1 = 114
    k2 = 1014
    k3 = 10203
    o = np.cos(k1*x) + np.cos(k1*y)
    o += 1.5*(np.cos(k2*x) + np.cos(k2*y))
    o += 2*(np.cos(k3*x) + np.cos(k3*y))
    r2  = (x-0.5)**2+(y-0.5)**2
    o += 10*np.exp(-2*np.sqrt(r2))
    return o**2

```

```python
# sample the function with QTCI

xlim = [0.,1.]
ylim = [0.,1.]

# create the interpolator
IP = interpolate.Interpolator(f,tol=1e-2,nb=10,xlim=xlim,ylim=ylim,dim=2)



# check the evaluated points as a reference
rse,zse = IP.get_evaluated()
print(len(rse),"evaluations")
rse = np.array(rse)


# evaluate the function and the interpolation
nn = 100
xs = np.linspace(xlim[0],xlim[1],nn,endpoint=False)
ys = np.linspace(ylim[0],ylim[1],nn,endpoint=False)
zs,zsi = [],[]

for y in ys:
    for x in xs:
        zsi.append(IP(x,y))
        zs.append(f(x,y))
```

```python
# now plot the data
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.title("Interpolated")
plt.imshow(np.array(zsi).reshape(nn,nn),vmax=np.max(zsi),cmap="inferno")
plt.xticks([]) ; plt.yticks([]) ; plt.xlabel("kx") ; plt.ylabel("ky")
plt.colorbar(ticks=[],label="LDOS",orientation="horizontal")

plt.subplot(1,3,2)
plt.title("Sampled points")
rse = np.array(rse)
plt.scatter(rse[:,0],-rse[:,1],c=zse,s=0.1,cmap="inferno",vmax=np.max(zse))
plt.xticks([]) ; plt.yticks([]) ; plt.xlabel("kx") ; plt.ylabel("ky")
plt.axis("equal") ; plt.axis("off")
plt.colorbar(ticks=[],label="LDOS",orientation="horizontal")


plt.subplot(1,3,3)
plt.title("Reference image")
plt.imshow(np.array(zs).reshape(nn,nn),vmax=np.max(zsi),cmap="inferno")
plt.xticks([]) ; plt.yticks([]) ; plt.xlabel("kx") ; plt.ylabel("ky")
plt.colorbar(ticks=[],label="LDOS",orientation="horizontal")


plt.tight_layout()
```

```python
# evaluate the function and the interpolation
nn = 100
xs = np.linspace(0.45,0.55,nn,endpoint=False)
ys = np.linspace(0.45,0.55,nn,endpoint=False)
zs,zsi = [],[]

for y in ys:
    for x in xs:
        zsi.append(IP(x,y))
        zs.append(f(x,y))
        

# now plot the data
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Interpolated")
plt.imshow(np.array(zsi).reshape(nn,nn),vmax=np.max(zsi),cmap="inferno")
plt.xticks([]) ; plt.yticks([]) ; plt.xlabel("kx") ; plt.ylabel("ky")
plt.colorbar(ticks=[],label="LDOS",orientation="horizontal")

plt.subplot(1,2,2)
plt.title("Reference image")
plt.imshow(np.array(zs).reshape(nn,nn),vmax=np.max(zsi),cmap="inferno")
plt.xticks([]) ; plt.yticks([]) ; plt.xlabel("kx") ; plt.ylabel("ky")
plt.colorbar(ticks=[],label="LDOS",orientation="horizontal")


plt.tight_layout()
```
